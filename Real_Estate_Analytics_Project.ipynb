# 3. Customer Preferences and Behavior


Problem:
Understanding and predicting customer preferences for property features, locations, and price ranges is essential for effective marketing and sales strategies.

Data Analysis Approach:

Analyze customer data, including search patterns, inquiries, and past purchases.
Utilize clustering and segmentation techniques to identify distinct customer profiles and preferences.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Sample Data (Replace this with your actual data loading)
data = {
    'Customer_ID': range(1, 101),
    'Interest_Rate': np.random.rand(100),
    'Government_Policy_Index': np.random.rand(100),
    'Property_Condition': np.random.choice(['Good', 'Average', 'Poor'], 100),
    'Location': np.random.choice(['Urban', 'Suburban', 'Rural'], 100),
    'Market_Demand': np.random.rand(100),
    'Recent_Sales_Price': np.random.rand(100),
    'Customer_Searches': np.random.rand(100) * 100,
    'Rental_Income': np.random.rand(100),
    'Area_sqft': np.random.rand(100) * 2000,
    'Customer_Preferences': np.random.choice(['Low', 'Medium', 'High'], 100),
    'Property_Type': np.random.choice(['Apartment', 'House', 'Condo'], 100),
    'Market_Value': np.random.rand(100) * 500000
}

df = pd.DataFrame(data)

# Handle missing values
df.fillna(method='ffill', inplace=True)

# Encode categorical variables
le = LabelEncoder()
df['Property_Condition'] = le.fit_transform(df['Property_Condition'])
df['Location'] = le.fit_transform(df['Location'])
df['Customer_Preferences'] = le.fit_transform(df['Customer_Preferences'])
df['Property_Type'] = le.fit_transform(df['Property_Type'])

# Scale numerical variables
scaler = StandardScaler()
numerical_features = ['Interest_Rate', 'Government_Policy_Index', 'Market_Demand', 'Recent_Sales_Price',
                      'Customer_Searches', 'Rental_Income', 'Area_sqft', 'Market_Value']
df[numerical_features] = scaler.fit_transform(df[numerical_features])

# Select relevant features for clustering
selected_features = ['Interest_Rate', 'Government_Policy_Index', 'Market_Demand', 'Recent_Sales_Price',
                     'Customer_Searches', 'Rental_Income', 'Area_sqft', 'Market_Value']
# Apply K-Means Clustering
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(df[selected_features])

# PCA for dimensionality reduction and visualization
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
principal_components = pca.fit_transform(df[selected_features])
df['PCA1'] = principal_components[:, 0]
df['PCA2'] = principal_components[:, 1]

# Plot clusters based on PCA components
plt.figure(figsize=(10, 7))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df, palette='viridis', s=100, alpha=0.8)
plt.title('Customer Segmentation')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()
plt.figure(figsize=(8, 5))
sns.countplot(x='Customer_Preferences', data=df, palette='viridis')
plt.title('Customer Preferences')
plt.xlabel('Preference Level')
plt.ylabel('Count')
plt.show()
plt.figure(figsize=(10, 6))
sns.countplot(x='Property_Type', hue='Location', data=df, palette='viridis')
plt.title('Property Type Distribution by Location')
plt.xlabel('Property Type')
plt.ylabel('Count')
plt.legend(title='Location')
plt.show()
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Market_Demand', y='Recent_Sales_Price', hue='Cluster', data=df, palette='viridis', s=100, alpha=0.8)
plt.title('Market Demand vs. Recent Sales Price')
plt.xlabel('Market Demand')
plt.ylabel('Recent Sales Price')
plt.show()

```
